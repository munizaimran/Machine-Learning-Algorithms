```python
#import libaraies
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
from sklearn.metrics import accuracy_score #to check accuracy of thr model


#import dataset and display it
dataset = pd.read_csv('xyz.csv')
dataset

#to discover the shape of dataset(row * column)
dataset.shape

#for statistical analysis
dataset.describe()

#data cleaning (this removes all the null values in the dataset)
dataset.isnull().any()
dataset = dataset.fillna(method='ffill')

#plotting dataset
#'x' and 'y' are the column names


dataset.plot(x='x', y='y', style='o')  
plt.title('title')  
plt.xlabel('x-axis')  
plt.ylabel('y-axis')  
plt.show()

#slicing the data
X = data.iloc[:, 0:]
y = data.iloc[:,-1] to get the last column, which is usually target


#FOR MULTIPLE LINEAR REGRESSION
X = dataset[['column_a', 'column_b', 'column_c', 'column_d', 'column_e', 'column_f']].values
y = dataset['target'].values

#spliting the dataset into train and test
#20% test, 80% train
#random state splits the data randomly, giving you a different result every time
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#training the algorithm by importing LinearRegression class and calling the fit() method
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

#to make predictions
y_pred = regressor.predict(X_test)

#comparing actual output (y_test) with predicted output(y_pred)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df

#to visualize the first 25 results
df1 = df.head(25)
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()

#to plot the line of best fit
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()

#check accuracy
print(accuracy_score(y_test, y_pred))

#for statistical output
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
